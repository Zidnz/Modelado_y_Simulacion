```python
import random

n = 15
M = 50.0
L = 200.0


total_ventas_mes = 0
total_clientes_mes = 0
total_compradores_mes = 0

print(f"{'Día':<5} | {'Clientes':<10} | {'Compradores':<12} | {'Ventas ($)':<10}")
print("-" * 50)


for dia in range(1, 31):
    clientes_hoy = random.randint(n, n + 10)

    compradores_hoy = 0
    ventas_hoy = 0

    for _ in range(clientes_hoy):
        if random.random() <= 0.60:
            compradores_hoy += 1
            precio_producto = random.uniform(M, L)
            ventas_hoy += precio_producto

    total_clientes_mes += clientes_hoy
    total_compradores_mes += compradores_hoy
    total_ventas_mes += ventas_hoy


    print(f"{dia:<5} | {clientes_hoy:<10} | {compradores_hoy:<12} | {ventas_hoy:<10.2f}")

print("-" * 50)
print("RESUMEN MENSUAL")
print(f"Clientes totales del mes:           {total_clientes_mes}")
print(f"Clientes que compraron en el mes:   {total_compradores_mes}")
print(f"Ventas totales del mes:            ${total_ventas_mes:.2f}")
```
