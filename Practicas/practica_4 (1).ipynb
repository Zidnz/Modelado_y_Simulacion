**Planteamiento del problema**

■Una empresa vende productos durante 30 días.

■Se sabe que:

•Cada día llegan entre 40 y 70 clientes

•El 65% de los clientes compra

•Cada cliente compra 1 producto

•Cada producto cuesta entre $120 y $250

•La empresa inicia con 200 productos en inventario y cada 6 días resurtir los 200

•Si el producto se termina, se pierde la venta

Lo que debe calcular el programa

1.Ventas por día

2.Inventario restante por día

3.Cuántas ventas se perdieron

4.Monto perdido

5.Ganancia total

6.Día con mayor venta

7.Día con menor venta

```python
import random
import pandas as pd
```

```python
dias = 10
clientes_rango = (40, 70)
probalidad_compra = 0.65
precio_rango = (120, 250)
stock_inicial = 200
frecuencia_resurtida = 10
```

```python
inventario_actual = stock_inicial
historial = []

for dia in range(1, dias + 1):
    if dia > 1 and (dia - 1) % frecuencia_resurtida == 0:
        inventario_actual = stock_inicial

    clientes_hoy = random.randint(*clientes_rango)

    quieren_comprar = 0
    for _ in range(clientes_hoy):
        if random.random() <= probalidad_compra:
            quieren_comprar += 1

    ventas_reales = min(quieren_comprar, inventario_actual)
    perdidas_unidades = quieren_comprar - ventas_reales

    ingresos_hoy = sum(random.uniform(*precio_rango) for _ in range(ventas_reales))

    inventario_actual -= ventas_reales

    historial.append({
        "Día": dia,
        "Clientes": clientes_hoy,
        "Ventas_OK": ventas_reales,
        "Ventas_Perdidas": perdidas_unidades,
        "Ingresos_$": round(ingresos_hoy, 2),
        "Stock_Final": inventario_actual
    })
```

```python
df = pd.DataFrame(historial)
print(df.to_string(index=False))

print("═" * 60)
print("═" * 22, "RESUMEN MENSUAL", "═" * 21)
print("═" * 60)
print(" " * 60)
ganacia_total = df["Ingresos_$"].sum()
dia_max = df.loc[df["Ingresos_$"].idxmax()]
dia_min = df.loc[df["Ingresos_$"].idxmin()]

print(f"Ganancia Total del Mes: ${ganacia_total:,.2f}")
print("═" * 60)
print(" " * 60)
print(f"Total de ventas perdidas: {df['Ventas_Perdidas'].sum()} unidades")
print("═" * 60)
print(" " * 60)
print(f"Día con MAYOR venta: Día {int(dia_max['Día'])} (${dia_max['Ingresos_$']})")
print("═" * 60)
print(" " * 60)
print(f"Día con MENOR venta: Día {int(dia_min['Día'])} (${dia_min['Ingresos_$']})")
print("═" * 60)
```
